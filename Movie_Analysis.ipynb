# Question 1: What kinds of properties are associated with most and less successful movies?
To answer this question, I will first calculate the financial benefit realized by each movie by subtracting the budget adjusted (accounting for inflation over time) from the revenue adjusted. We say that there is a profit when revenue generated from a movie exceeds the expenses, costs, and taxes involved in sustaining the activity in question, and vice versa for losses.


# Calculate Movies profit or loss

```python

# calculate profit/ loss(losses when there are negative values) of each movie, by substracting the budget_adj from the revenue_adj of each movie
df['profit']=df['revenue_adj']-df['budget_adj']

#show dataset first two rows
df.head(2)
```

# Create a function to extract the min and the max values
So as to facilitate the answer to some questions, i will create a function that willextract the min and the max simultaneously, by using the idxmin()/idxmax() function to get the row label of the minimum/maximum value

```python
# Create a minmax function by using idxmin()/idxmax() function to get the row label of the minimum/maximum value.

def min_max(x):
    # calculate maximum value
    highest = df[x].idxmax()
    highest_details = pd.DataFrame(df.loc[highest])
    
    # calculate minimum value
    lowest = df[x].idxmin()
    lowest_details = pd.DataFrame(df.loc[lowest])  
    
    print('Movie which has highest '+x+' : ', df['original_title'][highest])
    print('Movie which has lowest '+x+' : ', df['original_title'][lowest])
    
    return pd.concat([highest_details, lowest_details], axis=1)
```

# Which genres are most popular and profitable, overall and overtime?

## Number of movies produced per genre

```python


# split the genres  from combined value into single genre value 
df_split = df.assign(genres = df['genres'].str.split('|')).explode('genres')

df_split.head(2)
```

```python



df_split['genres'].value_counts(ascending = False).plot(kind='bar',figsize = (10,8),color='g');

plt.title('Movies by Genre, 1960-2015', size=18)
plt.xlabel('Genres', size=12)
plt.ylabel('Number of occurance')
sns.set_style("whitegrid")
```

As we can see, Drama movies are the most frequent movie genre. The top 5 dominant movie genres all over these years (1960 - 2015) are Drama, Comedy,Thriller, Action and Adventure.

# Which genres are more profitable overall?

```python
# Calculate profitability of each genre
profit_and_genres_df = df_split[['genres', 'profit']]

# Group by genres and calculate the mean profit for each genre
mean_profit_vs_genre_df = profit_and_genres_df.groupby('genres', as_index=False)['profit'].mean()

# Sort the result based on profit in descending order
mean_profit_vs_genre_df = mean_profit_vs_genre_df.sort_values(by='profit', ascending=False)

# Display the top genres by mean profit in a table format
mean_profit_vs_genre_df.head()

```

```python
#plot
mean_profit_vs_genre_df['profit'].plot(stacked=False, kind='barh', figsize = (10,8),color='g');
plt.title('Genres classification by profit', size=18)
plt.xlabel('profit made', size=12);
sns.set_style("whitegrid")
```

Even that, "Drama" genres are the most frequent genre produced overall, but the most profitable one is "Animation". So the top 5 most profitable genres of movies from 1960 to 2015 are : Animation, Adventure, Family, Fantasy, Science Fiction.



# Which genres are more profitable from year to year?


In this part, i will focus on the top 5 most profitable genres found in the last part to investigate which is the more profitable genre from year to year.

```python
#plot 5 most profitable genres over time

genres_peryear_df = df_split[['original_title', 'profit', 'genres','release_year','popularity']]

genre_profit_df = genres_peryear_df.groupby(['genres', 'release_year'], as_index=False)['profit'].mean()
fig,ax = plt.subplots(figsize=(10,8))

for genres in ['Animation','Adventure','Family', 'Fantasy', 'Science Fiction']:
    ax.plot(genre_profit_df[genre_profit_df.genres==genres].release_year,
            genre_profit_df[genre_profit_df.genres==genres].profit,label=genres)

ax.set_xlabel("release year")
ax.set_ylabel("profit")
plt.title('Profit made by genres from 1960 to 2015')
plt.show()
```

Most profitable Genres varies quite a lot from year to year and have known an upward in recent years. In 2015, Science fiction is the most profitable in 2015 followed by adventure in the second position and family in the third position.

# What is the evolution of the genres according to popularity from year to year?

Same conclusion for the Most popular Genres. They varie quite a lot from year to year and have known a significant uptrend in recent years.

By selected genres,Science fiction is still the most popular in 2015 followed by adventure in the second position and action in the third position.
